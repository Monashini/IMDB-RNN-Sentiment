```python
import tensorflow
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    "I loved the movie",              # Positive
    "The acting was great",           # Positive
    "What a boring film",             # Negative
    "I will never watch this again",  # Negative
    "Absolutely fantastic experience",# Positive
    "Worst direction and plot",       # Negative
    "Such a beautiful performance",   # Positive
    "The story made no sense",        # Negative
    "Highly recommended to all",      # Positive
    "It was a waste of time"          # Negative
]
```

```python
sent
```

```python
#define the vocab size
voc_size=10000
```

```python
##ohe rep
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr ## these are index where the word is located 
```

```python
##word embedding layer
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## at the end in rnn we send fixed number of words in a sentences with its timestamp so we use pad _seq
sent_length=8 #so max wrd size is 8 so make em all 8 itll add 0 either in pre or post ofthe senteces
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
##feature repr ___ to make it into that dim for the word embeddong
dims=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dims,input_length=sent_length))
model.compile('adam','mse')
model.summary()
```

```python
model.predict(embedded_docs) ##so alll wrds in sentences is rep in 10 dim vector
#foreach wrd in a sentence id gotrep as 10d vector
```
